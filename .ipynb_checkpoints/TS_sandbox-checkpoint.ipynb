```python
import pandas as pd
import hmmlearn as hmm
import tslearn
from tslearn.clustering import TimeSeriesKMeans
import numpy
import matplotlib.pyplot as plt
from tslearn.utils import to_time_series_dataset
```

```python
poor = pd.read_csv("C:/Users/met48/Desktop/Schelling_ML-main/Schelling_ML-main/bank_reserves_outputs_poor.csv", header=None)
middle = pd.read_csv("C:/Users/met48/Desktop/Schelling_ML-main/Schelling_ML-main/bank_reserves_outputs_middle.csv", header=None)
rich = pd.read_csv("C:/Users/met48/Desktop/Schelling_ML-main/Schelling_ML-main/bank_reserves_outputs_rich.csv", header=None)
```

```python
poor
```

```python
seed=0
X_train = to_time_series_dataset(poor)
sz = X_train.shape[1]

km = TimeSeriesKMeans(n_clusters=10, verbose=True, random_state=seed)
y_pred = km.fit_predict(X_train)

plt.figure()
for yi in range(10):
    plt.subplot(3, 3, yi + 1)
    for xx in X_train[y_pred == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(km.cluster_centers_[yi].ravel(), "r-")
    plt.xlim(0, sz)
    #plt.ylim(-4, 4)
    plt.text(0.55, 0.85,'Cluster %d' % (yi + 1),
             transform=plt.gca().transAxes)
    if yi == 1:
        plt.title("Euclidean $k$-means")
```

```python
seed=0
X_train = to_time_series_dataset(rich)
sz = X_train.shape[1]

km = TimeSeriesKMeans(n_clusters=10, verbose=True, random_state=seed)
y_pred = km.fit_predict(X_train)

plt.figure()
for yi in range(10):
    plt.subplot(4, 4, yi + 1)
    for xx in X_train[y_pred == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(km.cluster_centers_[yi].ravel(), "r-")
    plt.xlim(0, sz)
    plt.ylim(0, 250)
    plt.text(0.55, 0.85,'Cluster %d' % (yi + 1),
             transform=plt.gca().transAxes)
    if yi == 1:
        plt.title("Euclidean $k$-means")
```

```python
import numpy as np
from hmmlearn import hmm
X1 = np.random.random((10, 2))  # 2-D
X2 = np.random.random((5, 2))   # must have the same ndim as X1.
X = np.concatenate([X1, X2])
lengths = np.array([len(X1), len(X2)])
model = hmm.GaussianHMM(algorithm='viterbi', covariance_type='diag', covars_prior=0.01,
      covars_weight=1, init_params='stmc', means_prior=0, means_weight=0,
      min_covar=0.001, n_components=2, n_iter=10, params='stmc',
      random_state=None, startprob_prior=1.0, tol=0.01, transmat_prior=1.0,
      verbose=False)
```

```python
model.fit(X, lengths)
Z = model.predict(X)
states = pd.unique(Z)
```
