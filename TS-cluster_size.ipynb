```python
import pandas as pd
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer
import matplotlib.pyplot as plt
```

# Bank Reserves

```python
br_pca = pd.read_csv('extracted_features/bank_reserves_pca_standardized.csv')
br_pca = br_pca.drop('Unnamed: 0', axis=1)
```

```python
br_dae = pd.read_csv('extracted_features/bank_reserves_dae.csv', header=None)
```

```python
br_dcae = pd.read_csv('extracted_features/bank_reserves_dcae.csv', header=None)
```

```python
def elbow_plot_kmeans(min_k, max_k, df, title):
    Sum_of_squared_distances = []
    K = range(min_k, max_k)
    for k in K:
        km = KMeans(n_clusters=k)
        km = km.fit(df)
        Sum_of_squared_distances.append(km.inertia_)
    plt.plot(K, Sum_of_squared_distances, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Sum_of_squared_distances')
    plt.title(title)
    plt.show()
```

```python
elbow_plot_kmeans(1, 15, br_pca, "PCA Features")
```

```python
elbow_plot_kmeans(1, 15, br_dae, "DAE Features")
```

```python
elbow_plot_kmeans(1, 15, br_dcae, "DCAE Features")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4, 5]:
    km = KMeans(n_clusters=i)
    q, mod = divmod(i, 2)
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(br_pca) 
```
